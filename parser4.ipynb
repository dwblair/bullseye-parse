```python
def plotter(filename,spectrum,color):
    with open(filename) as data_file:
        data = json.load(data_file)
    x=data["spectrometer"]["wavelengths"]
    y=data["spectra"][spectrum]
    pylab.plot(x, y,color=color,label=filename.split('-')[6].split('.')[0])
    #pylab.title(spectrum)
    pylab.xlabel('nm')
```

```python
filename=sparklist[0]
spectrum='absorbance'
with open(filename) as data_file:
    data=json.load(data_file)
x=data["spectrometer"]["wavelengths"]
y=data["spectra"][spectrum]
dat=[]
for i in range(0,len(x)):
    dat.append((x[i],y[i]))
```

```python
df = pd.DataFrame(data = dat, columns=['wavelength', 'absorbance'])
df
```

```python
df.to_csv('test.csv',index=False,header=False)
```

```python
type='FLAME_NIR'
sparklist=glob.glob('data2/*'+type+'*.json')
for filename in sparklist:
    plotter(filename,'absorbance',color='red')
pylab.title(type)
type='FLAME_NIR'
sparklist=glob.glob('carrots/*'+type+'*.json')
for filename in sparklist:
    plotter(filename,'absorbance',color='blue')
pylab.title(type)
type='FLAME_NIR'
sparklist=glob.glob('carrots2/*'+type+'*.json')
for filename in sparklist:
    plotter(filename,'absorbance',color='green')
pylab.title(type)
type='FLAME_NIR'
sparklist=glob.glob('carrots3/*'+type+'*.json')
for filename in sparklist:
    plotter(filename,'absorbance',color='yellow')
pylab.title(type)
type='FLAME_NIR'
sparklist=glob.glob('carrots4/*'+type+'*.json')
for filename in sparklist:
    plotter(filename,'absorbance',color='orange')
pylab.title(type)
```
