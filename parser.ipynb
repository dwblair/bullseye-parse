```python
%matplotlib inline  
import os
import json
import pylab
import numpy as np
```

```python
def plotter(filename,spectrum):
    with open(filename) as data_file:
        data = json.load(data_file)
    x=data["spectrometer"]["wavelengths"]
    y=data["spectra"][spectrum]
    pylab.plot(x, y,label=filename.split('-')[6].split('.')[0])
    #pylab.title(spectrum)
    pylab.xlabel('nm')
```

```python
width = 14
height = 6
pylab.figure(figsize=(width, height))
files=os.listdir('data')
for file in files:
    filename='data/'+file
    plotter(filename,'sampleLight')
pylab.legend()
```
